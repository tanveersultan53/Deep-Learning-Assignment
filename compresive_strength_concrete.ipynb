```python
#spliting traing and testing and validation data
train, validate, test = \
              np.split(df.sample(frac=1, random_state=42), 
                       [int(.6*len(df)), int(.8*len(df))])
```

**Scaling
Standardizing the data i.e. to rescale the features to have a mean of zero and standard deviation of 1.**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
#spliting traind and target value
#trianin Data

X_train = train.iloc[:,:-1]         # Features - All columns but last
y_train = train.iloc[:,-1]          # Target - Last Column
#validation Data
X_val = validate.iloc[:,:-1]         # Features - All columns but last
y_val = validate.iloc[:,-1]          # Target - Last Column
#testing Data
X_test = test.iloc[:,:-1]         # Features - All columns but last
y_test = test.iloc[:,-1]          # Target - Last Column
```

```python
#Now Tranfer The Data into Scalar 
X_train = sc.fit_transform(X_train)
X_val = sc.fit_transform(X_val)
X_test=sc.fit_transform(X_test)
```

```python
#Define the Network Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.activations import relu , softmax
```

```python
model = Sequential()
model.add(Dense(10 , activation='relu' , input_shape=(8,)))
model.add(Dense(8))
model.add(Dense(6))
model.add(Dense(1 , activation="linear"))
```

```python
model.summary()
```

```python
#compilation Setp
model.compile(optimizer="rmsprop" , loss="mse", metrics=['mse'], )
```

```python
history = model.fit(X_train , y_train , validation_data=(X_val , y_val) , epochs=100 , batch_size=32)
```

```python
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validate'], loc='upper left')
plt.show()
```

```python
#Evaluate the Model
model.evaluate(X_test,y_test, batch_size=10)
```
